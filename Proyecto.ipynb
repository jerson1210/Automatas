```python
from IPython.display import display, HTML
from sys import exit
#e
def calcular_imc(peso, altura):
    return peso / (altura ** 2)

def pedir_datos():
    while True:
        try:
            peso = float(input('Ingresa tu peso en kg: '))
            altura = float(input('Ingresa tu altura en metros: '))
            break
        except ValueError:
            print('Por favor ingresa solo números')
    return peso, altura

def verificar_datos():
   while True:
        respuesta = input('¿Son estos tus datos reales? (s/n): ')
        if respuesta.lower() == 's':
            return False
        elif respuesta.lower() == 'n':
            return True
        else:
            print('Por favor ingresa s o n')

def verificar_peso(imc):
    if imc < 15 or imc > 40:
        print('Por favor visita a un médico antes de comenzar cualquier programa de entrenamiento')
        exit()

def asignar_rango(imc):
    if imc < 15:
        return 0
    elif imc >= 15 and imc < 18:
        return 1
    elif imc >= 18 and imc < 25:
        return 2
    elif imc >= 25 and imc < 30:
        return 3
    elif imc >= 30 and imc < 40:
        return 4
    else:
        return 5

def elegir_entrenamiento():
    opciones_entrenamiento = {'hipertrofia': 5, 'resistencia': 6, 'fuerza': 7}
    while True:
        entrenamiento = input('¿Qué tipo de entrenamiento deseas hacer? (hipertrofia/resistencia/fuerza): ')
        if entrenamiento.lower() in opciones_entrenamiento:
            return opciones_entrenamiento[entrenamiento.lower()]
        else:
            print('Por favor elige una opción válida')

def elegir_dia():
    dias_semana = {'lunes': 8, 'martes': 9, 'jueves': 1, 'viernes': 2}
    while True:
        dia = input('¿Qué día de la semana vas a hacer ejercicio? (lunes/martes/jueves/viernes): ')
        if dia.lower() in dias_semana:
            return dias_semana[dia.lower()]
        else:
            print('Por favor elige una opción válida')

#Luego, se verifica si el día ingresado por el usuario está en el diccionario dias_semana.
#Si el día está en el diccionario, la función devuelve el valor asociado a ese día (por ejemplo, si el usuario ingresa “lunes”, la función devuelve 8).
#Si el día no está en el diccionario, se muestra un mensaje de error y el bucle continúa ejecutándose, pidiendo al usuario que ingrese nuevamente un día



def volver_atras():
    while True:
        respuesta = input('¿Deseas volver atrás y cambiar tus elecciones? (s/n): ')
        if respuesta.lower() == 's':
            return True
        elif respuesta.lower() == 'n':
            return False
        else:
            print('Por favor ingresa s o n')

peso, altura = pedir_datos()
imc = calcular_imc(peso, altura)
print(f'Tu IMC es: {imc:.2f}')

while verificar_datos():
    peso, altura = pedir_datos()
    imc = calcular_imc(peso, altura)
    print(f'Tu IMC es: {imc:.2f}')

verificar_peso(imc)

rango_imc = asignar_rango(imc)
entrenamiento = elegir_entrenamiento()
dia = elegir_dia()

palabra = f'{rango_imc}{entrenamiento}{dia}0'
palabras = [palabra]
print(f'Tus opciones son: Rango IMC: {rango_imc}, Entrenamiento: {entrenamiento}, Día: {dia}')
print(f'La palabra generada es: {palabra}')

while volver_atras():
    rango_imc = asignar_rango(imc)
    entrenamiento = elegir_entrenamiento()
    dia = elegir_dia()

    palabra = f'{rango_imc}{entrenamiento}{dia}0'
    palabras.append(palabra)
    print(f'Tus opciones son: Rango IMC: {rango_imc}, Entrenamiento: {entrenamiento}, Día: {dia}')
    print(f'La palabra generada es: {palabra}')

if palabras[-1].endswith('0'):
    palabras[-1] = palabras[-1][:-1]
    #La línea if palabras[-1].endswith('0'): verifica si el último elemento de la lista

palabra_final = ''.join(palabras)
print(f'La palabra final es: {palabra_final}')



# Supongamos que ya tienes el enlace almacenado en una variable llamada 'download_url'
download_url1 = "https://docs.google.com/document/d/1nh4rzcIBqCNDklWxIhtkJzvCRq-kBbSOCNZ122PwCxc/edit"
link_text1 = "imc 1 lunes"
download_url2 = "https://docs.google.com/document/d/1-99v-NK-RMmZ1-OPjxMH8PeX2KIpbh5nFQgQ4kFiC3E/edit"
link_text2 = "imc 1 martes"
download_url3 = "https://docs.google.com/document/d/1OjSV3OuN_DuuieUuLIgC6o5_dZJfnBrhbhwEoXFGc3E/edit"
link_text3 = "imc 1 jueves"
download_url4 = "https://docs.google.com/document/d/1YTXVeZbftzp0Hd1B7FEMxTQuwKqv-6hSMqpXor07lo0/edit"
link_text4 = "imc 1 viernes"


download_url5 = "https://docs.google.com/document/d/1pcqSzgy0eL_i4GyPL5Eig4yFoyWQJIdcYHOOU_pNms0/edit"
link_text5 = "imc 2 lunes"
download_url6 = "https://docs.google.com/document/d/1iImpJK-mdID_zEUc3E1x4QbmoZ59-viTim02dPxUD_g/edit"
link_text6 = "imc 2 martes"
download_url7 = "https://docs.google.com/document/d/1e5R3_fiR1tUxpvlCR9jpm_YWRrpWXdfxYhEJuUyCBM8/edit"
link_text7 = "imc 2 jueves"
download_url8 = "https://docs.google.com/document/d/1wcqDZfb3gEdIpWWmuB8JYRsM-e0EN2j8y_gSHAYDBdg/edit"
link_text8 = "imc 2 viernes"

download_url9 = "https://docs.google.com/document/d/14KBM40VS_kJ0DmSYclpqp8k1dv_Rls4AK8m7BlhjOKc/edit"
link_text9 = "imc 3 lunes"
download_url10 = "https://docs.google.com/document/d/1mvLW1hGQyV9Bj4yEKDJqFnEnvJ49LENDT4DaqkjOLIQ/edit"
link_text10 = "imc 3 martes"
download_url11 = "https://docs.google.com/document/d/1K6EOjenUqrgZxLT3Mj7E_UGAJMkw35KAZ5PsGvP4e9I/edit"
link_text11 = "imc 3 jueves"
download_url12 = "https://docs.google.com/document/d/1cQodaLP-TF8u-SP00iNWgI4uOTJhsG0JW7TFqO3eXIM/edit"
link_text12 = "imc 3 viernes"

download_url13 = "https://docs.google.com/document/d/1GS4K4IZWde1ktD8Al4DF5IcarwOGh-w2uYRiHBVUd7g/edit"
link_text13 = "imc 4 lunes"
download_url14 = "https://docs.google.com/document/d/1KmHZuSUUvOniQVymFlpjGKZgInhhTxqWXUbvOu20lSg/edit"
link_text14 = "imc 4 martes"
download_url15 = "https://docs.google.com/document/d/1M9So3BqGXMH8j4kW_QYUvVBTxm4mlFpi7d5Q4KxM_hA/edit"
link_text15 = "imc 4 jueves"
download_url16 = "https://docs.google.com/document/d/1-ajRm-9yHFOohGVfbIwZMEe19bcBK_D2bw60wrw5N6w/edit"
link_text16 = "imc 4 viernes"


# Genera el código HTML para el enlace
link_html1 = f'<a href="{download_url1}" target="_blank">{link_text1}</a>'
link_html2 = f'<a href="{download_url2}" target="_blank">{link_text2}</a>'
link_html3 = f'<a href="{download_url3}" target="_blank">{link_text3}</a>'
link_html4 = f'<a href="{download_url4}" target="_blank">{link_text4}</a>'

link_html5 = f'<a href="{download_url5}" target="_blank">{link_text5}</a>'
link_html6 = f'<a href="{download_url6}" target="_blank">{link_text6}</a>'
link_html7 = f'<a href="{download_url7}" target="_blank">{link_text7}</a>'
link_html8 = f'<a href="{download_url8}" target="_blank">{link_text8}</a>'

link_html9 = f'<a href="{download_url9}" target="_blank">{link_text9}</a>'
link_html10 = f'<a href="{download_url10}" target="_blank">{link_text10}</a>'
link_html11 = f'<a href="{download_url11}" target="_blank">{link_text11}</a>'
link_html12 = f'<a href="{download_url12}" target="_blank">{link_text12}</a>'

link_html13 = f'<a href="{download_url13}" target="_blank">{link_text13}</a>'
link_html14 = f'<a href="{download_url14}" target="_blank">{link_text14}</a>'
link_html15 = f'<a href="{download_url15}" target="_blank">{link_text15}</a>'
link_html16 = f'<a href="{download_url16}" target="_blank">{link_text16}</a>'



# Muestra el enlace en el notebook


def verificar_palabra_en_automata(d, palabra_final):
    try:
        return d.accepts_input(palabra_final)
    except ValueError:
        return False
d = p()


es_palabra_valida = verificar_palabra_en_automata(d, palabra_final)
if es_palabra_valida:
    if rango_imc == 1 and palabras[-1].endswith('8') :
        display(HTML(link_html1))
    elif rango_imc == 1 and palabras[-1].endswith('9') :
        display(HTML(link_html2))
    elif rango_imc == 1 and palabras[-1].endswith('1') :
        display(HTML(link_html3))
    elif rango_imc == 1 and palabras[-1].endswith('2') :
       display(HTML(link_html4))

    elif rango_imc == 2 and palabras[-1].endswith('8') :
        display(HTML(link_html5))
    elif rango_imc == 2 and palabras[-1].endswith('9') :
        display(HTML(link_html6))
    elif rango_imc == 2 and palabras[-1].endswith('1') :
        display(HTML(link_html7))
    elif rango_imc == 2 and palabras[-1].endswith('2') :
        display(HTML(link_html8))

    elif rango_imc == 3 and palabras[-1].endswith('8') :
        display(HTML(link_html9))
    elif rango_imc == 3 and palabras[-1].endswith('9') :
        display(HTML(link_html10))
    elif rango_imc == 3 and palabras[-1].endswith('1') :
        display(HTML(link_html11))
    elif rango_imc == 3 and palabras[-1].endswith('2') :
       display(HTML(link_html12))

    elif rango_imc == 4 and palabras[-1].endswith('8') :
       display(HTML(link_html13))
    elif rango_imc == 3 and palabras[-1].endswith('9') :
        display(HTML(link_html14))
    elif rango_imc == 3 and palabras[-1].endswith('1') :
        display(HTML(link_html15))
    elif rango_imc == 3 and palabras[-1].endswith('2') :
       display(HTML(link_html16))


    else:
        # Muestra un mensaje si la palabra es válida pero rango_imc no está entre 1 y 4
        print('La palabra es válida en el autómata pero rango_imc no está entre 1 y 4.')

```
